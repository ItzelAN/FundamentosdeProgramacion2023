#**Operadores aritméticos**

---
####Los **operadores aritméticos** se utilizan con valores numéricos para realizar operaciones matemáticas comunes. A continuación se enlistan todos los operadores aritméticos:

image.png

####Se puede asignar cualquier operación a una variable. Esta nos servirá para almacenar el resultado en caso de que lo necesitemos en otra parte del código. 


```python
#Operaciones aritméticas asignadas a variables
Suma = 19 + 87 + 40 + 3
Resta = 50 - 17
Multiplicacion = 7 * 9
Division = 85 / 4
Division_piso = 85 // 4
Modulo = 25 % 2
Potencia = 5 ** 3

print(f"Suma = {Suma}\nResta = {Resta}\nMultiplicación = {Multiplicacion}")
print(f"Division = {Division}\nDivision de piso = {Division_piso}")
print(f"Modulo = {Modulo}\nPotencia = {Potencia}")
```

####**Operaciones aritméticas directas en `print()`**
####Tiene la desventaja de que el resultado no se guarda para futuras operaciones. Si no se requiere guardar es una buena opción emplear a `print()` para ver el resultado de una operación.

```python
#Operaciones aritméticas directas en print()
print(f"Suma = {19 + 87 + 40 + 3}\nResta = {50 - 17}\nMultiplicación = {7 * 9}")
print(f"Division = {85 / 4}\nDivision de piso = {85 // 4}")
print(f"Modulo = {25 % 2}\nPotencia = {5 ** 3}")
```

####En Python existe la posibilidad de hacer operaciones rápidas y verificar cuál es su resultado sin necesidad de asignarlo a variable o de agregar un print(). Cabe mencionar que esto funcionará si lo único que queremos observar es ese resultado y no realizar ningún otro proceso.

```python
#Ejemplo de operación rápida
19 + 87 + 40 + 3
```

# **Operadores de asignación**

---

####Los **operadores de asignación** se utilizan para asignar valores a variables.

image.png

```python
#Ejemplo con los operadores de asignación
x = 15 #La variable x recibe el valor de 5
print("Valor inicial de x = ",x)
x += 3 #Asignación con suma
print("Valor suma de x = ",x)
x -= 7 #Asignación con resta
print("Valor resta de x = ",x)
x *= 4 #Asignación con multiplicación
print("Valor multiplicación de x = ",x)
x /= 2 #Asignación con división
print("Valor división de x = ",x)
x %= 4 #Asignación con módulo
print("Valor módulo de x = ",x)
x **= 9 #Asignación con potencia
print("Valor potencia de x = ",x)
x //= 6 #Asignación con división de piso
print("Valor división de piso de x = ",x)
```

#**Operadores de comparación (Relacionales)**

---
####Los **operadores relacionales** comparan los valores de dos variables o funciones y retornan un resultado Verdadero (**True**) o Falso (**False**) según la condición.

image.png


```python
#Ejemplo que solicita 2 números y aplica todos los operadores relacionales
x = int(input("Ingrese el primer número: "))
y = int(input("Ingrese el segundo número: "))

#La operación se realiza dentro de print() también se puede asignar a variable
print(f"\n{x} es igual a {y} = {x == y}")
print(f"{x} es diferente de {y} = {x != y}")
print(f"{x} es mayor que {y} = {x > y}")
print(f"{x} es menor que {y} = {x < y}")
print(f"{x} es mayor o igual que {y} = {x >= y}")
print(f"{x} es menor o igual que {y} = {x <= y}")
```

#**Operadores lógicos**

---
####Los **operadores lógicos** se utilizan para tomar una decisión basada en múltiples condiciones. Los operadores lógicos utilizados en Python son `and`, `or` y `not`.

image.png


```python
#Ejemplo utilizando el operador lógico and
Danae = 43
Victor = 39
Gael = 51

#Ambas condiciones son verdaderas
print("¿Danae es mayor que Víctor y menor que Gael?")
print("Respuesta = ", Danae > Victor and Danae < Gael)

#Solo una de las condiciones es verdadera
print("\n¿Danae es menor que Víctor y menor que Gael?")
print("Respuesta = ", Danae < Victor and Danae < Gael)

#Ninguna condición es verdadera
print("\n¿Danae tiene la misma edad de Víctor y de Gael?")
print("Respuesta = ", Danae == Victor and Danae == Gael)
```

```python
#Ejemplo utilizando el operador lógico or

#Ambas condiciones son verdaderas
print("¿Danae es mayor que Víctor ó menor que Gael?")
print("Respuesta = ", Danae > Victor or Danae < Gael)

#Solo una de las condiciones es verdadera
print("\n¿Danae es menor que Víctor ó menor que Gael?")
print("Respuesta = ", Danae < Victor or Danae < Gael)

#Ninguna condición es verdadera
print("\n¿Danae tiene la misma edad de Víctor ó de Gael?")
print("Respuesta = ", Danae == Victor or Danae == Gael)
```

```python
#Ejemplo utilizando el operador lógico not
#La condición es verdadera
print("¿Danae no es mayor que Víctor y menor que Gael?")
print("Respuesta = ", not(Danae > Victor and Danae < Gael))

#La condición es falsa
print("\n¿Danae no tiene la misma edad de Víctor ó de Gael?")
print("Respuesta = ", not(Danae == Victor or Danae == Gael))
```

# **Operadores de pertenencia**

---
####Los **operadores de pertenencia** se utilizan para comprobar si una secuencia se presenta en un objeto.

image.png


```python
#Ejemplo utilizando los operadores de pertenencia en una cadena
frase = "¡Hola mundo! Mi nombre es Itzel."

#Verificamos si se encuentra una subcadena dentro de la frase
print(f"¿La cadena \"Itzel\" esta en la frase \"{frase}\" ?")
print("Respuesta = ", "Itzel" in frase)
print(f"\n¿La cadena \"Perro\" esta en la frase \"{frase}\" ?")
print("Respuesta = ", "Perro" in frase)

#Verificamos si no se encuentra una subcadena dentro de la frase
print(f"\n¿La cadena \"Itzel\" no esta en la frase \"{frase}\" ?")
print("Respuesta = ", "Itzel" not in frase)
print(f"\n¿La cadena \"Perro\" no esta en la frase \"{frase}\" ?")
print("Respuesta = ", "Perro" not in frase)
```
