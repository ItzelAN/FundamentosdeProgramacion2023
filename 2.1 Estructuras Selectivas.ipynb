# **Estructura selectiva `if`**

---
####La estructura selectiva `if` ejecuta las instrucciones sólo si se cumple una condición (verdadera). Si la condición es falsa, no se hace nada.

####**Sintaxis:**


```
if condición:
  Instrucción(es)
```
####Las líneas de código después de los dos puntos y con sangría (Identación) solo serán ejecutadas cuando la **condición if** sea **Verdadera**. Las tareas finalizarán cuando el código no contenga sangría.





####**Ejemplo 1**

> Determine si un número dado es positivo.



````
num = 5
if num > 0:
  print(f"El número {num} es positivo")
````

# **Estructura selectiva `if - else`**

---
####La estructura selectiva `if - else` decide qué instrucciones ejecutar dependiendo si una condición es verdadera o falsa.

####**Sintaxis:**


```
if condición:
  Instrucción(es)
else:
  Instrucción(es)
```
####**Nota:** Se debe recordar que `else` no lleva condición ya que se ejecuta automáticamente cuando la condición evaluada es falsa.






####**Ejemplo 1**

> Determine si un número dado es positivo o negativo.



````
num = -5
if num > 0:
  print(f"El número {num} es positivo")
else:
  print(f"El número {num} es negativo")
````

# **Estructura selectiva `if-elif-else`**

---
####La sentencia `if-elif-else` depende de dos o más condiciones, que son evaluadas en orden. La primera que es verdadera determina qué instrucciones serán ejecutadas.

####**Sintaxis:**

```
if condición:
  Instrucción(es)
elif condición:
  Instrucción(es)
else:
  Instrucción(es)
```
####**Nota:**






####**Ejemplo 1**

> Determine si un número dado es positivo, negativo o cero.



````
num = 0
if num == 0:
  print(f"El número {num} es cero.")
elif num > 0:
  print(f"El número {num} es positivo.")
else:
  print(f"El número {num} es negativo.")
````

# **Estructura selectiva `match - case`**

---
####Una sentencia `match` recibe una expresión y compara su valor con patrones sucesivos dados en uno o más bloques `case`. Sólo se ejecuta el primer patrón que coincide y también es capaz de extraer componentes (elementos de una secuencia o atributos de un objeto) de un valor y ponerlos en variables.

####**Sintaxis:**

```
match parámetro:
  case Primero:
    Instrucción(es) primero
  case Segundo:
    Instrucción(es) segundo
  case _:
    Instrucción(es) si no se encuentra una coincidencia
```

````
dato = 1

match dato:
  case 1:
    print("Bienvenido")
  case 2:
    print("Hola")
  case _: #Caso predeterminado
    print("Adiós")
````

# **Operador ternario**

---



####El **operador ternario** o expresión condicional en Python nos permite acceder a una de dos opciones a partir de una condición.

####El operador ternario en Python tiene tres partes claramente diferenciadas:

1.   La condición, que llamaremos `condición`.
2.   La expresión izquierda, que llamaremos `izquierda`, y que se evalúa si la condición es `True` y se ignora si es `False`.
3.   La expresión derecha, que llamaremos `derecha`, y que se evalúa si la condición es `False` y se ignora si es `True`.

#### **Sintaxis:**

```
izquierda if condición else derecha
```

####La diferencia con un `if` estándar es que el operador ternario no es una estructura de control que permite dirigir el flujo del programa, sino que actúa como un operador que define una expresión.

####**Ejemplo 1**

> Uno de los usos más directos del operador es hacer que una variable tome un valor u otro en función de una condición. Necesitamos que una variable `X` tome un valor de **'Par'** si el valor de otra variable `Y` es par o que tome el valor de **'Impar'** si el valor de `Y` es impar.


````
#Ejemplo con if - else sin usar el operador ternario
Y = 15
if Y % 2 == 0:
  X = 'Par'
else:
  X = 'Impar'
print(f'El número {Y} es {X}.')
````

````
#Ejemplo con el operador ternario
Y = 14
X = 'Par' if Y % 2 == 0 else 'Impar'
print(f'El número {Y} es {X}.')
````

# **Anidamiento**


---



####El concepto de anidamiento es muy utilizado en la programación y consiste en insertar (**anidar**) una estructura dentro de otra. La expresión *“si anidados”* se refiere a que podemos utilizar una sentencia si (**if**) dentro de otra sentencia si (**if**). Esto se emplea cuando se tienen que cumplir varias condiciones para poder ejecutar una acción.

####**Sintaxis:**



```
if condición_1:
  instrucciones
else:
  if condición_2:
    instrucciones
    if condición_3:
      instrucciones
    else:
      instrucciones
  else:
    instrucciones
    if condición_4:
      instrucciones
    else:
      instrucciones

```
####Uno de los aspectos más confusos de las sentencias si (if) anidados es saber qué **else** se asocia con qué si (if). La computadora evalúa la expresión condicional de arriba abajo. Tan pronto como la computadora encuentra una condición verdadera, ejecuta la instrucció (es) y salta el resto de la escalera. Si no es verdad ninguna condición, la computadora ejecutará el else final. Éste actúa frecuentemente como condición por defecto; es decir si fallan todas las pruebas anteriores, la computadora realiza la última instrucción(es) else. Si no hay else final y son falsas las condiciones, no se hace nada.
