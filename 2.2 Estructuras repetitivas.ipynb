# **Estructura repetitiva `while`**

---
####La estructura repetitiva `while` ejecuta repetidamente una serie de instrucciones mientras que se cumpla una condición (verdadera). El bloque de código seguirá ejecutándose hasta que la condición regrese un valor Falso.

####**Sintaxis:**


```
while condición:
  Instrucción(es)
```




####**Ejemplo 1**

> Mostrar todos los números que se encuentran del 1 al 10. 



````python
num = 1 #Variable de control
print("---Inicio---\nLos números del 1 al 10 son:")
while num <= 10:
  print(num)
  num += 1
print("---Fin---")
````

#**Instrucción** `break`


---




####En Python, la instrucción `break` le proporciona la oportunidad de terminar con la ejecución de los bucles `while` o `for` cuando se activa una condición externa. Debe poner la instrucción `break `dentro del bloque de código bajo la instrucción de su bucle, generalmente después de una instrucción `if` condicional.

####**Ejemplo 2**


> Romper un ciclo indefinido `while True` por medio de la instrucción `break`. El ciclo disminuirá un número de 1 en 1 hasta llegar a 3.


````python
#Este tipo de ciclo correrá indefinidamente hasta que se le aplique el caso con break
x = 10
print("---Inicio---")
while True:
    print(x)
    x -= 1
    if x == 3: #El ciclo terminará cuando x sea 3
      break #La instrucción break termina el ciclo
print("---Fin---")
````

#**Instrucción** `continue`


---



####En Python, La instrucción `continue` da la opción de omitir la parte de un bucle `while` o `for` en la que se activa una condición externa, pero continuar para completar el resto del bucle. Es decir, la **iteración actual del bucle se interrumpirá**, pero el programa **volverá a la parte superior del bucle**.

####**Ejemplo 3**


> Mostrar los números del 1 al 20 exceptuando a los múltiplos de 3.

````python
num = 0 #Variable de control
print("---Inicio---")
while num < 20:
   num+=1
   if num % 3 == 0:
      continue #La instrucción continue continua con la siguiente iteración
   print(num)
print("---Fin---")
````

#**Estructura** `while - else`

---


####La estructura `while` en **Python** puede tener una cláusula `else` asociada:

####**Sintaxis:**


```
while condición:
  Instrucción(es)
else:
  Instrucción(es)
```
####Esto le permite generar una opción alternativa cuando la condición se vuelve falsa.


####**Ejemplo 4**


> Determinar si el usuario acierta o no una adivinanza. Se le darán 3 intentos para acertar.



````python
intento = 3
while intento >= 1:
  print(f"\n¿Qué animal tiene las 5 vocales?\nTienes {intento} intentos para acertar.")
  respuesta = input("\nRespuesta: ").capitalize()
  if respuesta == "Murciélago" or respuesta == "Murcielago":
    print(f"\n¡Felicidades! {respuesta} es la respuesta correcta.")
    break
  else:
    print("\nFallaste!")
    intento -= 1
else:
  print("\nHas agotado tus tres intentos")
````

# **Estructura repetitiva `for`**

---
####La estructura repetitiva `for` en **Python** se utiliza para iterar sobre una secuencia. En Python tenemos diversos tipos de datos que pueden fungir como secuencia:


*   Cadenas
*   Rangos
*   Listas
*   Conjuntos
*   Tuplas
*   Diccionarios

####El ciclo for ejecuta repetidamente una serie de instrucciones hasta que la secuencia termina.

####**Sintaxis:**


```
for variable_iterativa in secuencia:
  Instrucción(es)
```

####**Ejemplo 5**

> Recorrer e imprimir todas las letras de una palabra.


````python
for letra in "Python":
  print(letra)
````

####**Ejemplo 6**

> Buscar una letra en una palabra. El ciclo debe detenerse cuando se encuentre la letra que se busca.

````python
for letra in "Python":
  if letra == 'h':
    print("Se encontró la letra h\nFin del ciclo")
    break #Utilizamos a break para romper el ciclo for
  print(letra)
````

####**Ejemplo 7**

> Buscar una letra en una palabra. Cada que se encuentre a la letra esta no se imprimirá en pantalla.

````python
for letra in "Python":
  if letra == 'h':
    continue #Utilizamos a continue para saltar esa iteración
  print(letra)
````

#**Instrucción** `range()`


---
####La función `range()` devuelve una secuencia de números, empezando por 0 por defecto, e incrementa en 1 (por defecto), y se detiene antes de un número especificado.

#### **Sintaxis**

```
range(Inicio, Fin-1, Incremento / Decremento)
```
####Parámetros de `range()`:

**Inicio** ▶ *Opcional*
> Es un número entero que especifica en qué posición comenzar. Por defecto es 0

**Fin** ▶ *Requerido*
> Es un número entero que especifica en qué posición parar.

**Incremento / Decremento** ▶ *Opcional*
> Un número entero que especifica el incremento o decremento. Por defecto es 1.

####Normalmente se emplea a `range()` junto con el ciclo `for` para generar secuencias numéricas.

#**Sintaxis**

```
for variable_iterativa in range(Inicio, Fin-1, Incremento / Decremento):
  Instrucciones
```



####**Ejemplo 8**

> Mostrar todos los números que se encuentran del 1 al 10. 



````python
for num in range(1,11):
  print(num)
````

####**Ejemplo 9**

> Mostrar todos los números que se encuentran del 1 al 20 de aumentando de 3 en 3. 



````python
for num in range(1,21,3):
  print(num)
````
