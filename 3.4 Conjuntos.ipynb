#**Declarar un conjunto**

---

####Un conjunto es una colección no ordenada de **objetos únicos** (elementos no repetidos), los conjuntos son ampliamente utilizados en lógica y en matemáticas. Un conjunto se define en python de la siguiente forma:

####**Sintaxis:**

```
Conjunto = {Dato1, Dato2, ..., DatoN}
```

````
#Conjunto con objetos
conj = {5,3,2,6,1}
print("Objetos en el conjunto 1: ",conj)

#Conjunto vacío
conj_vacío = set()
print("Objetos en el conjunto 2: ",conj_vacío)
````

####**Método `set()`**

> El método `set()` crea un conjunto a partir de un objeto iterable.

````
#Conjunto vacío
conj_vacío = set()
print("Conjunto vacío: ",conj_vacío)

#Conjunto a partir de un rango
cnj = set(range(10))
print("Conjunto: ",cnj)
````

####**Tipos de dato**
> Al igual que en otras colecciones, los elementos de un conjunto pueden ser de diversos tipos:





````
c = {True, 3.14, False, "Hola mundo", 4, 'Z'}
c
````

> Sin embargo, un conjunto **no** puede incluir objetos mutables como **listas**, **diccionarios**, e incluso otros **conjuntos**:

````
c = {[1, 2]}
c
````

#**Agregar elementos al conjunto**


---


####Para agregar elementos al conjunto se pueden utilizar 2 diferentes funciones de Python:

*   Función **`add()`**

*   Función **`update()`**


####**Función `add()`**
> La función `add()` agrega un solo elemento al conjunto.

####**Sintaxis**



```
Conjunto.add(Elemento)
```

````
cnj = {1, 2, 3, 4, 6}
cnj.add(5) #Si el objeto ya esta en el conjunto no lo duplica
cnj
````

####**Función `update()`**

> La función actualizar `update()` agrega múltiples elementos a un conjunto a partir de un objeto iterable.

````
A = {1,2,3,4,5}
B = range(6,11)
C = [11,12,13,14,15]
print('Conjunto Original:    ',A)
A.update(B)
print('Conjunto Actualizado: ',A)
A.update(C)
print('Conjunto Actualizado: ',A)
````

#**Borrar elementos del conjunto**

---
####Para eliminar elementos del conjunto se pueden utilizar 3 diferentes funciones de Python:

*   Instrucción **`discard()`**

*   Función **`remove()`**

*   Función **`pop()`**

*   Función **`clear()`**

####**Función `discard()`**
> La función `discard` elimina un elemento del conjunto, si no encuentra al elemento no hace nada.
####**Sintaxis**

```
conjunto.discard(Elemento)
```

````
cnj = {1548, 'Alfredo', True, 0.89, 3, 'b', 6 , False, 'A'}
cnj.discard('Alfredo')
cnj
````

####**Función `remove()`**
> La función `remove()` elimina un elemento del conjunto, si no encuentra al elemento marcará error.
####**Sintaxis**

```
conjunto.remove(Elemento)
```

````
cnj.remove(6)
cnj
````

####**Función `pop()`**
> La función `pop()` elimina un elemento del conjunto de forma arbitraria.
####**Sintaxis**

```
conjunto.pop()
```

````
cnj.pop()
cnj
````

####**Función `clear()`**
> La función `clear()` elimina a todos los elementos del conjunto dejándolo vacío.
####**Sintaxis**

```
conjunto.clear()
```

````
cnj = {1, 2, 3, 4, 5, 6}
cnj.clear()
cnj
````

#**OPERACIONES BÁSICAS CON CONJUNTOS**


---


####Las principales operaciones que se producen con conjuntos son:
*   Unión
*   Intersección
*   Diferencia
*   Diferencia simétrica
*   Igualdad

#**Unión**


---


####La unión de dos conjuntos A y B es el conjunto A∪B que contiene todos los elementos de A y de B.

````
#Ejemplo 1 unión de conjuntos
A = {1, 2, 3, 4} #C_U={1,2,3,4,5,6}
B = {3, 4, 5, 6}
A.union(B) #Función
````

````
#Ejemplo 2 unión de conjuntos
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
c = {5, 6, 7, 8}
a | b | c #Operador
````

#**Intersección**


---


####La intersección de dos conjuntos A y B es el conjunto A∩B que contiene todos los elementos comunes de A y B.

````
#Ejemplo 1 intersección de conjuntos
A = {1, 2, 3, 4} #C_I={3,4}
B = {3, 4, 5, 6}
A.intersection(B) #Función
````

````
#Ejemplo 2 intersección de conjuntos
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
a & b #Operador
````

#**Diferencia**


---



####La diferencia entre dos conjuntos A y B es el conjunto A∖B que contiene todos los elementos de A que no pertenecen a B.

````
#Ejemplo 1 diferencia de conjuntos
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}
A.difference(B) #Función A - B
````

````
#Ejemplo 2 diferencia de conjuntos
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
b - a #Operador
````

#**Diferencia simétrica**


---


####La diferencia simétrica entre dos conjuntos A y B regresa otro conjunto con todos los elementos que se encuentran en A y B pero que no se encuentran en ambos.

````
#Ejemplo 1 diferencia simétrica de conjuntos
A = {1, 2, 3, 4} #C_DS={1,2,5,6}
B = {3, 4, 5, 6}
A.symmetric_difference(B) #Función
````

````
#Ejemplo 2 diferencia simétrica de conjuntos
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
b ^ a #Operador
````

#**Igualdad**


---


####Dos conjuntos son iguales si y solo si contienen los mismos elementos (a esto se lo conoce como principio de extensionalidad).

````
A = {1, 2, 3, 4, 5}
B = {4, 1, 5, 3, 2}
A == B
````

#**OPERACIONES COMPLEMENTARIAS CON CONJUNTOS**


---


####Otras operaciones que se producen con conjuntos son:
*   Subconjunto
*   Superconjunto
*   Conjunto Disconexo

#**Subconjunto**


---


####Se dice que B es un subconjunto de A cuando todos los elementos de B pertenecen también a A.

````
A = {1, 2, 3, 4}
B = {2, 3}
B.issubset(A)
````

#**Superconjunto**


---


####Inversamente de lo mencionado con anterioridad, se dice que A es un superconjunto de B.

````
A = {1, 2, 3, 4}
B = {2, 3}
A.issuperset(B)
````

#**Conjunto disconexo**


---


####Un conjunto es disconexo respecto de otro si no comparten elementos entre sí.

````
A = {1, 2, 3, 4}
B = {5, 6, 7, 8}
A.isdisjoint(B)
````

#**Conjunto inmutable `frozenset()`**


---

####Un conjunto inmutable `frozenset()` es una implementación similar a un conjunto mutable `set()` pero sin cambios. Es decir, comparte todas las operaciones de conjuntos anteriormente explicadas a excepción de aquellas que implican alterar sus elementos (`add()`, `discard()`, `remove()`, etcétera).

````
A = frozenset({1, 2, 3})
B = frozenset({3, 4, 5})
A.union(B)
````

````
A.remove(1)
````

#**Cambio de conjunto a lista**


---


####Un conjunto `set()` puede ser convertido a una lista y viceversa.

````
#Conjunto a partir de una lista
lista = ['a','a','r','o','u','p','r']
conjL = set(lista)
print("Conjunto a partir de una lista: ",conjL)

#Lista a partir de un conjunto
conjunto = {2,3,489,12.09,3.2,0.45}
ListC = list(conjunto)
print("Lista a partir de un conjunto: ",ListC)
````

#**Membresía `in`**


---
Para determinar si un elemento pertenece a un conjunto, utilizamos la palabra reservada `in`.

````
cnj = {1, 2, 3, 4, 5, 6}
print(2 in cnj)
print(8 in cnj)
````

---


####Para copiar un conjunto se utiliza a la función `copy()`.

````
Animales = {'Oso','Perro','Avestruz','Camello','Hiena','Armadillo'}
Animales_copia = Animales.copy()
print(Animales_copia)
````

#**Máximo y mínimo**

---

####La función `max()` devuelve el elemento del conjunto con valor máximo.

####La función `min()` devuelve el elemento del conjunto con valor mínimo.

````
A = {154,4,3,2900,47,245664,5802757,152}
print('Conjunto: ',A)
print('Máximo del conjunto: ',max(A))
print('Mínimo del conjunto: ',min(A))
````
