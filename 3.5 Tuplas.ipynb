#**Declarar una tupla**

---
####Una **tupla** es un tipo de dato que contiene una colección de elementos ordenados e inmutables. Una tupla se genera a partir de separar los elementos por una coma y englobarlos entre paréntesis.

####**Sintaxis:**

```
Tupla = (Dato1, Dato2, ..., DatoN)
```


````
frutas = ('Pera', 'Manzana', 'Ciruela', 'Piña', 'Mango')
tupla_vacía = ()
print(frutas,'\n',tupla_vacía)
````

####**Método `tuple()`**

> El método `tuple()` crea una tupla a partir de un objeto iterable.

````
vacia = tuple()
letras = tuple('aeiou')
rango = tuple(range(1,11))
lista = tuple(['Perro', 1, False, 0.65, True, 'A'])
conjunto = tuple({8,9,7,6,5,4})

print("Tupla vacía: ", vacia)
print("Tupla a partir de una cadena: ", letras)
print("Tupla a partir de un rango:   ", rango)
print("Tupla a partir de una lista:  ", lista)
print("Tupla a partir de un conjunto:", conjunto)
````

#**Acceso a los elementos de la tupla**

---



####Al igual que con las listas los elementos de la tupla están indexados, el primer elemento tiene índice **[0]**, el segundo elemento tiene índice **[1]**, etc. Podemos a partir de su índice dividir, concatenar, multiplicar, etc. Para acceder a los valores en la tupla, se utilizan los corchetes junto con el índice o índices para obtener el valor disponible en ese índice.

####**Sintaxis:**

```
Tupla[índice]
```

````
# Ejemplo de acceso mediante los índices de la tupla
print(frutas[2]) # Acceso a la posición 0
print(letras[1:5]) # Acceso a las posiciones de la 1 a la 4   1 : n-1

# Ejemplo de acceso al último elemento de la tupla
print(rango[-1]) # El negativo nos indica el valor a la derecha de esa posición
print(lista[-2:]) # Extrae los últimos dos elementos de la tupla
````

####**Función `index()`**

####La función `index()` devuelve el índice o posición de un elemento de la tupla.

####**Sintaxis:**


```
tupla.index(Elemento)
```





````
print("Índice del elemento Manzana: ", frutas.index('Manzana'))
````

#**Cambiar los elementos de la tupla**


---

####Las tuplas son **inmutables**, lo que significa que no podemos cambiar, añadir o eliminar elementos una vez creada la tupla.

####Pero hay una solución. Puedes convertir la tupla en una lista, cambiar la lista y volver a convertir la lista en tupla.

````
nombres = ('Miguel', 'Vanessa', 'Gustavo')
nombres[0]='Miguel Ángel'
print('Nombres: ', nombres)
````

````
# Ejemplo para cambiar un elemento de la tupla
nombres_tupla = ('Miguel', 'Vanessa', 'Gustavo')
print("Nombres iniciales: ", nombres_tupla)

#Convierto la tupla a lista y realizó el cambio
nombres_lista = list(nombres_tupla)
nombres_lista[0]='Miguel Ángel'

#Convierto la lista a tupla y visualizo el resultado
nombres_tupla = tuple(nombres_lista)
print('Nombres finales:   ', nombres_tupla)
````

#**Agregar tupla a tupla**


---


####Puedes añadir tuplas a tuplas, así que si quieres añadir un elemento, (o muchos), crea una nueva tupla con el elemento(s), y añádelo a la tupla existente.:


####**Sintaxis**



```
tupla1 += tupla2
```

````
letras = ('A','B','C')
print("Tupla inicial: ", letras)
letras += ('D','E','F') #Añado una tupla a otra
print("Tupla final: ",letras)
````

#**Desempaquetar una tupla**

---
####Cuando creamos una tupla, normalmente le asignamos valores. Esto se llama **"empaquetar"** una tupla, ejemplo:

```
tupla = (Dato_1, Dato_2, Dato_3)
```
####**Python**, también se nos permite extraer los valores de nuevo en variables. Esto se llama **"desempaquetar"**, ejemplo:

####**Sintaxis**

```
(Var_1, Var_2, Var_3) = tupla
```

````
#Ejemplo de desempaquetado
frutas = ('Manzana', 'Plátano', 'Sandía')
(rojo, amarillo, verde) = frutas
print(f"{rojo}\n{amarillo}\n{verde}")
````

####**Desempaquetado con asterisco** `*`

####En el ejemplo anterior, podemos notar que el número de variables debe coincidir con el número de valores de la tupla. Si el número de variables es menor que el número de valores, puede añadir un **`*`** al nombre de la variable y los valores se asignarán a la variable como una lista, ejemplo:

```
tupla = (Dato_1, Dato_2, Dato_3, ..., Dato_n)
(Var_1, Var_2, *Var_3) = tupla
```



````
#Ejemplo de desempaquetado con asterisco
frutas = ('Manzana', 'Plátano', 'Sandía', 'Melón', 'Lima', 'Guayaba')
(rojo, amarillo, *verde) = frutas
print(f"{rojo}\n{amarillo}\n{verde}")
````


#**Operaciones con tuplas**


####En **Python** encontramos varios tipos de métodos que pueden aplicarse a las tuplas:

*   Longitud
*   Concatenación
*   Repetición
*   Membresía
*   Máximo y mínimo
*   Conteo

#**Longitud**

---



####El método `len()` devuelve el número de elementos que se encuentran en la tupla.

````
len(frutas)
````

#**Concatenación**


---


####La concatenación (+) enlaza dos o más tuplas y crea una nueva tupla a partir de esta unión.

````
a = (1,2,3)
b = (4,5,6)
c = a+b
print(c)
````

#**Repetición**

---



####La repetición (*) repite una misma tupla el número de veces que se le indique y genera una nueva tupla a partir de estos datos.

````
a = (1,2,3)
b = a*3
print(b)
````

#**Membresía**


---


####La membresía `in` o `not in` evalúa si un elemento de la tupla es igual a otro elemento externo y devuelve un valor verdadero o falso.

````
frutas = ('Manzana', 'Plátano', 'Sandía', 'Melón', 'Lima', 'Guayaba')

if "Sandía" in frutas:
  print("La fruta se encuentra en la tupla.")
````

#**Máximo y Mínimo**


---


####La función `max()` devuelve el elemento de la tupla con valor máximo.

####La función `min()` devuelve el elemento de la tupla con valor mínimo.

````
num = (50, 8000, 465, 10300, 1, 0.86, 0.015)
print("Mínimo: ",min(num))
print("Máximo: ",max(num))
````

#**Conteo**


---
####La función `count()` devuelve el recuento de cuántas veces se produce un elemento en la tupla.

````
varios = (2,'a',1,2,'b',2,'c',1,'a',0.5,'a',0.7,2)
print("Conteo del número 2: ", varios.count(2))
````

#**Iteración de los elementos de una tupla**

---



````
#Iteración sobre los elementos de la tupla
frutas = ('Manzana', 'Plátano', 'Sandía', 'Melón', 'Lima', 'Guayaba')
for elemento in frutas:
  print(elemento)
````

````
#Iteración sobre los índices de la tupla
frutas = ('Manzana', 'Plátano', 'Sandía', 'Melón', 'Lima', 'Guayaba')
for indice in range(len(frutas)):
  print(indice)
````
