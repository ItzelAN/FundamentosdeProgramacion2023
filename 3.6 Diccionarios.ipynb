#**Declarar un diccionario**


---


####Un **diccionario** es una estructura de datos que permite almacenar cualquier tipo de dato como enteros, cadenas, listas e incluso otros diccionarios. Estos diccionarios nos permiten además identificar cada elemento por una clave (Key).

####**Sintaxis:**
```
Diccionario_Vacío = {}
Diccionario = {'Clave 1': Valor_Clave1, 'Clave 2': Valor_Clave2}
Diccionario_función = dict(Clave 1 = Valor_Clave1, Clave 2 = Valor_Clave2)
```



````
# Diccionario vacío
Dicc_vacío = {}
print(Dicc_vacío)

# Diccionario
Dicc = {'Nombre' : 'Itzel', 'Apellido Paterno': 'Aranguren', 'Apellido Materno': 'Navarro'}
print(Dicc)

#Diccionario por medio de la función dict
Dicc_func = dict(Nombre = 'Itzel', Apellido_Paterno = 'Aranguren', Apellido_Materno = 'Navarro')
print(Dicc_func)
````

####**Función `fromkeys()`**



####La función **`fromkeys()`** recibe como parámetros un iterable y un valor, devolviendo un diccionario que contiene como claves los elementos del iterable con el mismo valor ingresado. Si el valor no es ingresado, devolverá un valor none para todas las claves.

````
#Sin valor ingresado
Dicc_3 = dict.fromkeys(['a','b','c','d','e'])
print(Dicc_3)

#Con valor ingresado
Dicc_4 = dict.fromkeys(['a','b','c','d','e'],5)
print(Dicc_4)

#Diccionario a partir de un rango
Dicc_5 = dict.fromkeys(range(5),'Hola')
print(Dicc_5)
````

# **Acesso a los valores de un diccionario**


---

####Para acceder a los elementos del diccionario, se utilizarán corchetes junto con el identificador clave para obtener su valor **`['Clave 1']`**. Si la clave no está disponible, se marcará un error.

````
#Dicc = {'Nombre' : 'Itzel', 'Apellido Paterno': 'Aranguren', 'Apellido Materno': 'Navarro'}
print('\nNombre:', Dicc['Nombre'])
print('Apellido Paterno:', Dicc['Apellido Paterno'])
print('Apellido Materno:', Dicc['Apellido Materno'])
````

####**Función `get()`**

####También se puede utilizar a la función **`get()`** para acceder al valor de una clave. Si la clave no está disponible, devuelve el valor predeterminado Ninguno (none).

####**Sintaxis**
```
diccionario.get(Clave)
```



````
print('Nombre:', Dicc.get('Nombre'))
print('Edad:',Dicc.get('Edad'))
````

####La función **`get()`** a su vez permite definir un caso predeterminado si no se encuentra una clave.

````
print('Nombre:', Dicc.get('Nombre',"No se encuentra esa clave"))
print('Edad:',Dicc.get('Edad', "No se encuentra esa clave"))
````

####**Método `setdefault()`**

####El método **`setdefault()`** es similar a **`get()`**, pero este tiene dos funciones la primera es conocer el valor de una clave y la segunda es agregar una nueva clave y valor al diccionario.
####**Sintaxis**
```
diccionario.setdefault(Clave, Valor)
```



````
# Conocer el valor de una clave
print("Diccionario original: ", Dicc)
print('Nombre:', Dicc.setdefault('Nombre'))
print('Edad:', Dicc.setdefault('Edad'))
print("Diccionario Final: ", Dicc)

# Agregar una clave y valor al diccionario
print('Diccionario Inicial:', Dicc)
Dicc.setdefault('Año Nacimiento',1990)
print('Diccionario Final:', Dicc)
````

# **Agregar elementos o actualizar un diccionario**


---


####Puede actualizar un diccionario agregando una nueva entrada o un par clave-valor, a su vez, se puede modificar un valor existente en una clave.

````
print(Dicc)
Dicc['Nombre'] = 'Esteban' # Actualiza el valor de esa clave
Dicc['Género'] = 'Masculino' # Agrega una nueva clave con un valor
print(Dicc)
````

####**Función `update()`**

####La función **`update()`** recibe como parámetro otro diccionario. Si se tienen claves iguales, actualiza el valor de la clave repetida; si no hay claves iguales, este par clave-valor es agregado al diccionario.
####**Sintaxis**
```
diccionario.update(Otro diccionario)
```

````
Dicc_1 = dict(a = 1, b = 2, c = 3, d = 4)
Dicc_2 = dict(c = 5, d = 6, e = 7, f = 8)
print('Diccionario Inicial:', Dicc_1)
print('Segundo Diccionario:', Dicc_2)
Dicc_1.update(Dicc_2)
#Dicc_2.update(Dicc_1)
print('Diccionario Final:', Dicc_1)
print('Segundo Diccionario:', Dicc_2)
````

####**Función `zip()`**

####La función **`zip()`** recibe como parámetros dos elementos iterables, ya sea una cadena, una lista, etcétera. Ambos parámetros deben tener **el mismo número de elementos**. Se devolverá un diccionario relacionando el elemento i-esimo de cada uno de los iterables.

####**Sintaxis**


```
diccionario = dict(zip(Objeto iterable para claves, Objeto iterable para valores))
```



````
claves = ['Materia', 'NRC', 'Semestre']
valores = ['Programación', 40205, "Segundo Semestre"]
materia = dict(zip(claves, valores))
print(materia)
````

#**Iterar sobre un diccionario**


---
####Iterar sobre las **claves** y **valores** de un diccionario.

````
#Forma directa
Playera = dict(Cant_Playera = 20, Color_Playera = ['Rojo','Azul','Verde'],
               Tallas_Playera = ['XL','L','M','S','XS'], Precio = 157.83)

for clave in Playera:
  print(clave,':', Playera[clave]) #Extraemos la clave y su valor
````

````
#Utilizando el método items()
Playera = dict(Cant_Playera = 20, Color_Playera = ['Rojo','Azul','Verde'],
               Tallas_Playera = ['XL','L','M','S','XS'], Precio = 157.83)

for clave, valor in Playera.items():
  print(clave,':', valor) #Extraemos la clave y su valor
````

####Iterar sobre las **claves** de un diccionario.

````
#Forma directa
Playera = dict(Cant_Playera = 20, Color_Playera = ['Rojo','Azul','Verde'],
               Tallas_Playera = ['XL','L','M','S','XS'], Precio = 157.83)

for clave in Playera:
  print(clave) #Extraemos la clave
````

````
#Utilizando el método keys()
Playera = dict(Cant_Playera = 20, Color_Playera = ['Rojo','Azul','Verde'],
               Tallas_Playera = ['XL','L','M','S','XS'], Precio = 157.83)

for clave in Playera.keys():
  print(clave) #Extraemos la clave
````

####Iterar sobre los **valores** de un diccionario

````
#Forma directa
Playera = dict(Cant_Playera = 20, Color_Playera = ['Rojo','Azul','Verde'],
               Tallas_Playera = ['XL','L','M','S','XS'], Precio = 157.83)

for clave in Playera:
  print(Playera[clave]) #Extraemos la clave
````

````
#Utilizando el método values()
Playera = dict(Cant_Playera = 20, Color_Playera = ['Rojo','Azul','Verde'],
               Tallas_Playera = ['XL','L','M','S','XS'], Precio = 157.83)

for valor in Playera.values():
  print(valor) #Extraemos el valor
````

#**Diccionarios y tipos de datos admitidos**


---
####El diccionario puede contener cualquier tipo de dato incluyendo a las listas, tuplas, conjuntos e incluso otros diccionarios.

````
#Diccionario que guarda una lista, un conjunto y un diccionario
Varios = dict(Lista = ['Amarillo', 'Rojo', 'Verde'],
             Conjunto = {'Cuadrado', 'Triángulo', 'Círculo'},
             Diccionario = dict(Borrador = 5, Pluma = 2,
             Cuaderno = ['Raya', 'Cuadro', 'Dibujo'])
             )
print("Diccionario completo = ", Varios)
print("Diccionario con lista = ", Varios['Lista'])
print("Diccionario con conjunto = ", Varios['Conjunto'])
print("Diccionario con diccionario = ", Varios['Diccionario'])
print("Imprimir un dato de la lista = ", Varios['Lista'][2])
print("Verificar si existe un dato en el conjunto = ",'Cuadrado' in Varios['Conjunto'])
print("Imprimir un valor del diccionario = ", Varios['Diccionario']['Borrador'])
print("Imprimir los valores de una lista en el diccionario = ",Varios['Diccionario']['Cuaderno'])
print("Imprimir un sólo valor de una lista en el diccionario = ",Varios['Diccionario']['Cuaderno'][1])
````

####Iterar sobre los datos de un diccionario que contiene listas, conjuntos y diccionarios.
