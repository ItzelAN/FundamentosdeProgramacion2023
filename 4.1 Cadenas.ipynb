####El tipo de dato cadena (**string**) es áquel que se compone por secuencias de caracteres encerrados entre comillas que representan texto. Cada caracter de la cadena (incluidos los espacios en blanco) tiene un índice asignado. Si consideremos el orden de izquierda a derecha, el indice empieza en 0 para el primer carácter, etc. También se puede considerar el orden de derecha a izquierda, en cuyo caso al último carácter le corresponde el índice -1, al penúltimo -2 y así sucesivamente.

# **Operaciones con cadenas**


---


> Entre las operaciones que se pueden realizar con cadenas tenemos:

*   **Concatenación:** La concatenación une dos cadenas o más mediante el operador (+).


```
    "¡HOLA" + " " + "MUNDO!" -> ¡HOLA MUNDO!
```

*   **Repetición:** Cualquier cadena se puede repetir un número determinado de veces, mediante el operador (*).


```
    "¡HOLA MUNDO!" * 3 -> ¡HOLA MUNDO!¡HOLA MUNDO!¡HOLA MUNDO!
```




*   **División:** La división extrae cualquier parte de la cadena mediante el valor del índice o conjunto de índices de la cadena:

    ( [ : : ] -> [ Inicio : Fin(n-1) : Incremento / Decremento ] ).

```
    Frase = "¡HOLA MUNDO!"
    Frase[0] -> ¡
    Frase[-1] -> !
    Frase[2:] -> OLA MUNDO!
    Frase[2:8] -> OLA MU
    Frase[:8] -> ¡HOLA MU
    Frase[1::2] -> HL UD!
    Frase[:10:2] -> ¡OAMN
    Frase[::2] -> ¡OAMNO
    Frase[1:11:2] -> HL UD
    Frase[::] -> ¡HOLA MUNDO!
    Frase[::-1] -> !ODNUM ALOH¡
```






````python
Frase = "¡HOLA" + " " + "MUNDO!"
print(Frase)
print(Frase*3)
print(Frase[0])
print(Frase[-1])
print(Frase[2:])
print(Frase[2:8])
print(Frase[:8])
print(Frase[1::2])
print(Frase[:10:2])
print(Frase[::2])
print(Frase[1:11:2])
print(Frase[::])
print(Frase[::-1])
````

# **Cadenas multilínea**


---



> Se puede asignar una cadena de varias líneas a una variable usando tres comillas simples o dobles, al inicio y al final de la frase.

**Ejemplo:**

```
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''
```



````python
a = '''Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua.'''

b = """Lorem ipsum dolor sit amet,
consectetur adipiscing elit,
sed do eiusmod tempor incididunt
ut labore et dolore magna aliqua."""

print("Cadena multilínea con comillas simples:\n", a)
print("\nCadena multilínea con comillas dobles:\n",b)
````

# **Longitud**


---



> Para conocer la cantidad de caracteres y espacios en una cadena se utiliza la función **`len()`**.

**Ejemplo:**



```
Frase = "¡Hola mundo soy Itzel!"
print(f"La longitud de la frase '{Frase}' es: {len(Frase)}") -> La longitud de la frase '¡Hola mundo soy Itzel!' es: 22
```



````python
Frase = "¡Hola mundo soy Itzel!"
print(f"La longitud de la frase '{Frase}' es: {len(Frase)}")
````

# **Membresia**


---



> Se puede verificar si una letra o subcadena pertenece a otra cadena mediante el operador de membresía **`in`**.

####**Ejemplo:**


```
Frase = "¡Hola mundo soy Itzel!"
print("soy" in Frase) -> True
```



````python
Frase = "¡Hola mundo soy Itzel!"
print("soy" in Frase)
````



> El ejemplo anterior también se puede comparar dentro de una estructura if-else:



####**Ejemplo:**


```
Frase = "¡Hola mundo soy Itzel!"

if "soy" in Frase:
  print(f"La cadena 'soy' si se encuentra en el texto: {Frase}") -> La cadena 'soy' si se encuentra en el texto: ¡Hola mundo soy Itzel!
```



````python
Frase = "¡Hola mundo soy Itzel!"

if "soy" in Frase:
  print(f"La cadena 'soy' si se encuentra en el texto: {Frase}")
````



> No solo podemos verificar si una cadena se encuentra en el texto también podemos verificar casos en los que no se encuentre por medio de **`not in`**.



````python
Frase = "¡Hola mundo soy Itzel!"

if "Buen día" not in Frase:
  print(f"La cadena 'Buen día' no se encuentra en el texto: {Frase}")
````

# **Métodos para modificar cadenas**


---


> Existen diversos métodos para modificar cadenas. A continuación se describen algunos de estos métodos:

###**Método `upper()`**
> El método `upper()` cambia cualquier texto a mayúsculas.

**Ejemplo:**
```
Frase = "¡Hola mundo soy Itzel!"
print(Frase.upper()) -> ¡HOLA MUNDO SOY ITZEL!
```



````python
Frase = "¡Hola mundo soy Itzel!"
print(Frase.upper())
````

###**Método `lower()`**
> El método `lower()` transforma cualquier texto a minúsculas.

**Ejemplo:**
```
Frase = "¡Hola mundo soy Itzel!"
print(Frase.lower()) -> ¡hola mundo soy itzel!
```


````python
Frase = "¡Hola mundo soy Itzel!"
print(Frase.lower())
````

###**Método `title()`**
> El método `title()` transforma la primera letra de cada palabra del texto a mayúsculas.

**Ejemplo:**
```
Frase = "¡Hola mundo soy Itzel!"
print(Frase.title()) -> ¡Hola Mundo Soy Itzel!
```

````python
Frase = "¡Hola mundo soy Itzel!"
print(Frase.title())
````

###**Método `capitalize()`**
> El método `capitalize()` transforma la primera letra del texto a mayúsculas y las demás a minúsculas.

**Ejemplo:**
```
Frase = "hOlA"
print(Frase.capitalize()) -> Hola
```

````python
Frase = "hOlA MUNDO"
print(Frase.capitalize())
````

###**Método `swapcase()`**
> El método `swapcase()` transforma todas las letras mayúsculas del texto a minúsculas y todas las letras minúsculas del texto a mayúsculas.

####**Ejemplo:**
```
Frase = "hOlA"
print(Frase.swapcase()) -> HoLa
```

````python
Frase = "hOlA"
print(Frase.swapcase())
````

###**Método `strip()`**
> El método `strip()` elimina cualquier espacio en blanco que se encuentre al principio o al final de una cadena.

####**Ejemplo:**
```
Frase = "  ¡Hola mundo soy Itzel!  "
print(Frase.strip()) -> "¡Hola mundo soy Itzel!"
```

````python
Frase = "  Itzel Aranguren "
print("Con espacios en blanco:",Frase)
print("Sin espacios en blanco:",Frase.strip())
````

````python
#Este método también sirve para quitar caracteres o letras al inicio o al final
Frase = "__Itzel Aranguren**"
print("Con guión bajo y asteriscos:",Frase)
print("Sin guión bajo:",Frase.strip("_"))
print("Sin asteriscos:",Frase.strip("*"))
````

###**Método `replace()`**
> El método `replace()` busca una cadena dentro del texto y lo cambia por otra.

####**Ejemplo:**
```
Frase = "¡Hola mundo soy Itzel!"
print(Frase.replace("Itzel", "Ana")) -> ¡Hola mundo soy Ana!
```

````python
Frase = "¡Hola mundo soy Itzel!"
print(Frase.replace("Itzel", "Ana"))
````

###**Método `split()`**
> El método `split()`
devuelve una lista donde el texto entre el separador especificado en los paréntesis se convierte en los elementos de la lista.

####**Ejemplo:**
```
Frase = "¡Hola mundo soy Itzel!"
#Ejemplo separando a partir de los espacios en blanco
print(Frase.split()) -> ["¡Hola","mundo","soy","Itzel!"]
```

````python
Frase = "Hola Mundo; Mi Nombre; es Itzel"
#Separa la cadena a partir de ;
print(Frase.split(";"))
````

# **Métodos de escape**


---


> Existen diversos métodos de escape de caracteres para configurar la salida en pantalla. A continuación se describen algunos de estos métodos:

###**Salto de línea `\n`**
> Un salto de línea **\n** traslada cualquier texto o instrucción a una nueva línea.

####**Ejemplo:**
```
print("¡Hola\nmundo\nsoy\nItzel!") ->
¡Hola
mundo
soy
Itzel!
```



````python
print("¡Hola\nmundo\nsoy\nItzel!")
````

###**Sangría o Tabulación `\t`**
> La sangría o tabulación **\t** traslada cualquier texto hacia la derecha.

####**Ejemplo:**
```
print("¡Hola\tmundo\tsoy\tItzel!") -> ¡Hola	mundo	soy	Itzel!
```

````python
print("¡Hola\tmundo\tsoy\tItzel!")
````

###**Retroceso `\b`**
> El retroceso **\b** elimina un caracter o espacio en blanco anterior a la instrucción.

####**Ejemplo:**
```
print("¡Hola\bmundo\bsoy\bItzel!") -> ¡HolmundsoItzel!
```

````python
print("¡Hola \bmundo\bsoy Itzel!\b\b")
````

###**Diagonal invertida `\`**
> La diagonal invertida **\** nos permite mostrar cualquier caracter que ya este asociado a un instrucción específica en Python. Supongamos que se desea imprimir comillas " " en un texto y a su vez también se quiere mostrar una diagonal invertida \.

####**Ejemplo:**
```
Incorrecto
print("¡Hola "mundo" soy \ Itzel!") -> SyntaxError: invalid syntax

Correcto
print("¡Hola \"mundo\" soy \\ Itzel!") -> ¡Hola "mundo" soy \ Itzel!
```

````python
print("¡Hola "mundo" soy Itzel!")
````

````python
print("¡Hola \"mundo\" soy \\ Itzel!")
````

# **Otros métodos de cadenas**

---

###**Método `center()`**
> Centra un texto acorde a la cantidad de espacios que se le indique.

####**Ejemplo**
```
Frase = "¡Hola mundo soy Itzel!"

#Centra la frase impresa acorde a 100 espacios
print(Frase.center(100)) ->
                                       ¡Hola mundo soy Itzel!
#Centra la frase impresa acorde a 100 espacios y rellena los espacios con -
print(Frase.center(100, -)) ->
---------------------------------------¡Hola mundo soy Itzel!---------------------------------------
```



````python
Frase = "¡Hola mundo soy Itzel!"
# Centra la frase impresa acorde a 100 espacios
print(Frase.center(100))

#Centra la frase impresa acorde a 100 espacios
#Y rellena los espacios disponibles con -
print(Frase.center(100,"-"))
````

###**Método `count()`**
> Cuenta cuantas veces aparece una letra o secuencia de letras en la frase.


```
Frase = "¡Hola mundo soy Itzel!"

print(Frase.count('o')) -> 3
```



````python
Frase = "¡Hola mundo soy Itzel!"
print("La letra o aparece {} veces en el texto".format(Frase.count("o")))
````

###**Método `find()` e `index()`**
> Busca una letra o cadena en la frase y devuelve la posición donde se encontró dicho elemento, en cuanto encuentra una coincidencia se detiene el proceso.

La diferencia entre `find()` e `index()`, es que si buscamos un dato y esté no se encuentra, `find()` nos devolverá un valor -1 e `index()` nos generará una excepción.


```
Frase = "¡Hola mundo soy Itzel!"

print(Frase.find('mundo')) -> 6
print(Frase.index('mundo')) -> 6
```



````python
Frase = "¡Hola mundo soy Itzel!"

print("Índice donde se encontró la palabra mundo: ",Frase.find('mundo'))
print("Índice donde se encontró la palabra mundo: ",Frase.index('mundo'))
````

````python
# Dato inexistente mediante find()
Frase = "¡Hola mundo soy Itzel!"

print("Índice donde se encontró la palabra perro: ",Frase.find('perro'))
````

````python
# Dato inexistente mediante index()
Frase = "¡Hola mundo soy Itzel!"
# Generá una excepción por lo que habría que manejar un código por excepciones
print("Índice donde se encontró la palabra perro: ",Frase.index('perro'))
````
