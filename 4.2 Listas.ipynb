#**Declarar una lista**

---
####Una **lista** es un tipo de dato que contiene una colecci√≥n de elementos ordenados. Los elementos de una lista no necesitan ser del mismo tipo. Una lista se genera a partir de separar los datos de la lista por una coma y englobarlos entre corchetes.

####**Sintaxis:**

```
Lista = [Dato1, Dato2, ..., DatoN]
```


image.png

````
# Ejemplo de como definir las listas

fruta = ['Pera', 'Manzana', 'Ciruela'] # Lista de palabras (cadenas)
num = [1, 3, 5, 7, 9, 11] # Lista de n√∫meros enteros
Lista_combinada = ['A','B', 'C', 1, 2, 3]
Lista_vac√≠a = []

print(fruta,'\n',num,'\n',Lista_combinada,'\n',Lista_vac√≠a)
````

####**M√©todo `list()`**

> El m√©todo `list()` crea una lista a partir de un objeto iterable.

````
Vacia = list()
Vocales = list('aeiou')
Numeros = list(range(1,11))
Lista = list(['Perro', 1, False, 0.65, True, 'A'])

print("Lista vac√≠a: ", Vacia)
print("Lista vocales: ", Vocales)
print("Lista num√©rica: ", Numeros)
print("Lista: ", Lista)
````

#**Acceso a los elementos de la lista**

---



####Al igual que en los √≠ndices del tipo de dato cadena, los √≠ndices de la lista comienzan en **0** y se pueden dividir, concatenar, multiplicar, etc. Para acceder a los valores en las listas, se utilizan los corchetes junto con el √≠ndice o √≠ndices para obtener el valor disponible en ese √≠ndice.

####**Sintaxis:**

```
Lista[√≠ndice]
```

````
# Ejemplo de acceso mediante los √≠ndices de la lista

print(fruta[0]) # Acceso a la posici√≥n 0
print(num[1:5]) # Acceso a las posiciones de la 1 a la 4   1 : n-1
print(Lista_combinada[:4]) # Acceso a las posiciones de la lista hasta antes de 4
print(Lista_vac√≠a,'\n') # Se imprimen los valores de la lista

# Ejemplo de acceso al √∫ltimo elemento de la lista

print(fruta[-1]) # El negativo nos indica el valor a la derecha de esa posici√≥n
print(num[-2:]) # Extrae los √∫ltimos dos elementos de la lista
print(Lista_combinada[-5:-3],'\n')
````

####**Funci√≥n `index()`**

####La funci√≥n `index()` devuelve el √≠ndice o posici√≥n de un elemento de la lista.

####**Sintaxis:**


```
Lista.index(Elemento)
```





````
print("√çndice del elemento Manzana: ", fruta.index('Manzana'))
print("√çndice del elemento 5: ",num.index(5))
````

#**Cambiar los elementos de la lista**


---


####Para cambiar un elemento de la lista basta con entrar a la posici√≥n de ese elemento y asignarle un nuevo valor.

####**Sintaxis**



```
Lista[√≠ndice] = Nuevo_valor
```



````
# Ejemplo para cambiar un elemento de la lista

nombres = ['Miguel', 'Vanessa', 'Gustavo']
print('Nombres iniciales: ', nombres)
nombres[0]='Miguel √Ångel'
print('Nombres finales:   ', nombres,'\n')

# Ejemplo de cambio m√∫ltiple

nombres=['Miguel', 'Vanessa', 'Gustavo']
print('Nombres iniciales: ', nombres)
nombres[0:2]='Miguel √Ångel', 'Vanessa Monserrat'
print('Nombres finales:   ', nombres)
````

#**Agregar elementos a la lista**


---


####Para agregar elementos a una lista se pueden utilizar 3 diferentes funciones de Python:

*   Funci√≥n **`append()`**

*   Funci√≥n **`extend()`**

*   Funci√≥n **`insert()`**


####**Funci√≥n `append()`**
> La funci√≥n `append()` agrega un solo elemento al final de una lista.

####**Sintaxis**



```
Lista.append(Elemento)
```

````
letras = ['A','B','C','D','E']
print('Lista inicial: ', letras)
letras.append('F')
print('Lista final:   ', letras)
letras.append('G')
print('Lista final:   ', letras)
````

####**Funci√≥n `extend()`**
> La funci√≥n `extend()` agrega m√∫ltiples elementos al final de una lista a partir de un objeto iterable.

####**Sintaxis**



```
Lista.extend(Objeto_iterable)
```

````
lista = ['A','B','C','D','E']
print('Lista inicial: ', lista)
lista.extend(['F','G','H','I','J']) #Agrega elementos a partir de otra lista
print('Lista con lista:   ', lista)
lista.extend("ITZEL") #Agrega elementos a partir de una cadena
print('Lista con cadena:   ', lista)
lista.extend(range(5)) #Agrega elementos a partir de un rango
print('Lista con rango num√©rico:   ', lista)
````

####**Funci√≥n `insert()`**
> La funci√≥n `insert()` inserta un nuevo elemento en una posici√≥n espec√≠fica de la lista.
####**Sintaxis**

```
Lista.insert(√çndice, Elemento)
```

````
letras = ['A','B','C','D','E']
print('Lista inicial: ', letras)
letras.insert(1, 'a')
print('Lista final:   ', letras)
````

#**Borrar elementos de la lista**

---
####Para eliminar elementos de la lista se pueden utilizar 4 diferentes funciones de Python:

*   Instrucci√≥n **`del`**

*   Funci√≥n **`remove()`**

*   Funci√≥n **`pop()`**

*   Funci√≥n **`clear()`**

####**Funci√≥n `del`**
> La funci√≥n `del` elimina uno o varios elementos de la lista a partir de su posici√≥n.
####**Sintaxis**

```
del Lista[√çndice_a_borrar]
```

````
# Ejemplo 1 borrar un solo elemento de la lista
animales = ['Perro','Gato','Canario','Delf√≠n','Elefante']
print('Lista inicial: ', animales)
del animales[2]
print('Lista final:   ', animales, '\n')

# Ejemplo 2 borrar varios elementos de la lista
animales = ['Perro','Gato','Canario','Delf√≠n','Elefante']
print('Lista inicial: ', animales)
del animales[1:4]
print('Lista final:   ', animales)
````

####**Funci√≥n `remove()`**
> La funci√≥n `remove()` busca un elemento dado en la lista y elimina el primer elemento coincidente.
####**Sintaxis**

```
Lista.remove(Elemento)
```

````
# Ejemplo 1
animales = ['Perro','Gato','Canario','Delf√≠n','Elefante']
print('Lista inicial: ', animales)
animales.remove('Delf√≠n')
print('Lista final:   ', animales, '\n')

# Ejemplo 2
animales = ['Perro','Gato','Canario','Delf√≠n','Gato','Elefante']
print('Lista inicial: ', animales)
animales.remove('Gato')
print('Lista final:   ', animales, '\n')
````

####**Funci√≥n `pop()`**
> La funci√≥n `pop()` elimina y devuelve el elemento borrado de la lista por medio del √≠ndice.
####**Sintaxis**

```
Lista.pop(√çndice)
```

````
animales = ['Perro','Gato','Canario','Delf√≠n','Elefante']
print('Lista inicial: ', animales)
#Si no se indica el √≠ndice se borrara el √∫ltimo elemento de la lista
print('Elemento eliminado: ', animales.pop())
print('Lista final:   ', animales, '\n')
print('Elemento eliminado: ', animales.pop(2))
print('Lista final:   ', animales)
````

####**Funci√≥n `clear()`**
> La funci√≥n `clear()` elimina todos los elementos de una lista dej√°ndola vac√≠a.
####**Sintaxis**

```
Lista.clear()
```

````
print("Lista actual: ", animales)
animales.clear()
print('Lista de animales:   ',animales)
````
